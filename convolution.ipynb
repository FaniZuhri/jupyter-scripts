```python
import numpy as np
import matplotlib.pyplot as plt


def plot_signal(n, signal, title, color="b"):
    """Fungsi untuk menampilkan sinyal diskrit."""
    plt.stem(n, signal, linefmt=f"{color}-", markerfmt=f"{color}o", basefmt="k")
    plt.xlabel("n")
    plt.ylabel("Amplitude")
    plt.title(title)
    plt.grid()


# Definisi sinyal input (x[n]) dan respon impuls (h[n])
x = np.array([1, 2, 3, 4])
h = np.array([0, 1, 0.5])

# Konvolusi menggunakan np.convolve
y = np.convolve(x, h, mode="full")

# Plot hasil
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
n_x = np.arange(len(x))
plot_signal(n_x, x, "Input Signal x[n]", "b")

plt.subplot(3, 1, 2)
n_h = np.arange(len(h))
plot_signal(n_h, h, "Impulse Response h[n]", "g")

plt.subplot(3, 1, 3)
n_y = np.arange(len(y))
plot_signal(n_y, y, "Convolution Result y[n]", "r")

plt.tight_layout()
plt.show()
```
